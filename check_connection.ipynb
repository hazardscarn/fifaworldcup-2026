```python
import pandas as pd
from supabase import create_client, Client
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
def init_connection():
    url = os.getenv("SUPABASE_URL")
    key = os.getenv("SUPABASE_SECRET_KEY")
    return create_client(url, key)
```

```python
_conn = init_connection()
```

```python
def check_matches(_conn):
    try:
        
        result = _conn.table('worldcup_schedule').select("*").execute()
        # Check if the query was successful and if the result contains data
        if result and len(result.data) > 0:
            # The user is a member of the league
            return result
        else:
            # The user is not a member of the league
            return None
    except Exception as e:
        # Handle any potential exceptions (e.g., network errors, query errors)
        print(f"Error in check_user_membership: {e}")
        return False  # Return False in case of an error
```

```python
df = check_matches(_conn)
df_pd = pd.DataFrame(df.data)
df_pd.head()
```
